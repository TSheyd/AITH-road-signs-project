```python
from ultralytics import YOLO
```

```python
model = YOLO(path+"base_model/yolov8n.pt")
```

```python
model.info()
```

```python
history = model.train(
    data = path+"DATA/data.yaml",
    epochs = 200,
    imgsz = 320,
    device = 0,
    patience = 30,
    batch = 256,
    save_period = 20,
    verbose = True,
    optimizer = 'Adam',
    val = True,
    lrf = 0.001
)
```
